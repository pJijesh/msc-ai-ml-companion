## Gaussian Process regression with uncertainty


```python
# Gaussian Process regression with uncertainty
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, WhiteKernel

np.random.seed(42)

# True function
def f(x): return np.sin(2 * np.pi * x) + 0.5 * np.sin(4 * np.pi * x)

# Training data (sparse)
X_train = np.array([0.1, 0.3, 0.5, 0.7, 0.9]).reshape(-1, 1)
y_train = f(X_train.ravel()) + np.random.normal(0, 0.1, len(X_train))

# GP with RBF kernel + noise
kernel = RBF(length_scale=0.3) + WhiteKernel(noise_level=0.01)
gpr = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10, random_state=42)
gpr.fit(X_train, y_train)

X_test = np.linspace(0, 1, 300).reshape(-1, 1)
y_mean, y_std = gpr.predict(X_test, return_std=True)

fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(X_test, f(X_test.ravel()), 'g--', lw=2, label='True function')
ax.scatter(X_train, y_train, s=80, zorder=5, color='#1F3864', label='Training data')
ax.plot(X_test, y_mean, color='#2E75B6', lw=2, label='GP mean (prediction)')
ax.fill_between(X_test.ravel(), y_mean - 2*y_std, y_mean + 2*y_std,
                alpha=0.2, color='#2E75B6', label='95% confidence interval')
ax.set_xlabel('x'); ax.set_ylabel('f(x)')
ax.set_title('Gaussian Process Regression: Epistemic uncertainty grows away from training data')
ax.legend(); ax.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('ch07_gaussian_process.png', dpi=150, bbox_inches='tight')
plt.show()
print("Note how uncertainty (blue band) grows between training points — this is epistemic uncertainty.")
```
