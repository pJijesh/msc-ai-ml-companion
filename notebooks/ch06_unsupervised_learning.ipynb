## Find optimal k using silhouette score


```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Find optimal k using silhouette score
scores = []
K_range = range(2, 11)
for k in K_range:
    km = KMeans(n_clusters=k, n_init=10, random_state=42)
    labels = km.fit_predict(X)
    scores.append(silhouette_score(X, labels))

best_k = K_range[scores.index(max(scores))]
print(f"Best k by silhouette: {best_k}")
```

## PCA + t-SNE on MNIST (subset) for visualisation


```python
# PCA + t-SNE on MNIST (subset) for visualisation
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

# Load digits (1797 samples, 64 features — a mini MNIST)
digits = load_digits()
X = StandardScaler().fit_transform(digits.data)
y = digits.target

# PCA to 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# t-SNE to 2D (from PCA-50 for speed)
X_pca50 = PCA(n_components=50).fit_transform(X)
tsne = TSNE(n_components=2, perplexity=30, random_state=42, n_iter=1000)
X_tsne = tsne.fit_transform(X_pca50)

colors = plt.cm.tab10(np.linspace(0, 1, 10))
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

for digit in range(10):
    mask = y == digit
    axes[0].scatter(X_pca[mask,0], X_pca[mask,1], s=10, alpha=0.6,
                    color=colors[digit], label=str(digit))
    axes[1].scatter(X_tsne[mask,0], X_tsne[mask,1], s=10, alpha=0.6,
                    color=colors[digit], label=str(digit))

axes[0].set_title(f"PCA — explains {pca.explained_variance_ratio_.sum()*100:.1f}% variance", fontsize=11)
axes[0].legend(markerscale=2, title='Digit', bbox_to_anchor=(1.01,1))
axes[1].set_title("t-SNE — preserves local neighbourhoods (visualisation only)", fontsize=11)
axes[1].legend(markerscale=2, title='Digit', bbox_to_anchor=(1.01,1))

for ax in axes:
    ax.axis('off')

plt.suptitle('PCA vs t-SNE on the Digits Dataset', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('ch06_pca_tsne.png', dpi=150, bbox_inches='tight')
plt.show()
print("Remember: t-SNE axes have NO interpretable meaning.")
```
