## Timeline: Key Milestones

This chapter is primarily historical — no executable code. The cell below creates an interactive timeline of key AI milestones.

## Key AI milestones timeline (visualisation)


```python
# Key AI milestones timeline (visualisation)
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

milestones = [
    (1950, "Turing Test proposed"),
    (1956, "Dartmouth Conference\n(birth of AI)"),
    (1966, "ELIZA"),
    (1974, "First AI Winter begins"),
    (1980, "Expert Systems boom"),
    (1986, "Second AI Winter begins"),
    (1997, "Deep Blue beats Kasparov"),
    (2006, "Deep Belief Networks\n(Hinton)"),
    (2012, "AlexNet — deep learning\nrevolution begins"),
    (2017, "Transformer — Attention\nIs All You Need"),
    (2018, "BERT"),
    (2020, "GPT-3"),
    (2022, "ChatGPT / Stable Diffusion"),
    (2024, "EU AI Act in force"),
    (2026, "High-risk AI obligations\nenforced"),
]

fig, ax = plt.subplots(figsize=(14, 5))
years  = [m[0] for m in milestones]
labels = [m[1] for m in milestones]

ax.set_xlim(1945, 2030)
ax.set_ylim(-1, 2)
ax.axhline(y=0.5, color='#2E75B6', linewidth=2, alpha=0.5)
ax.axis('off')

for i, (yr, lbl) in enumerate(milestones):
    above = i % 2 == 0
    y_line = 0.9 if above else 0.1
    y_text = 1.2 if above else -0.4
    ax.plot([yr, yr], [0.5, y_line], color='#1F3864', linewidth=1.2, alpha=0.7)
    ax.plot(yr, 0.5, 'o', color='#2E75B6', markersize=7, zorder=5)
    ax.text(yr, y_text, lbl, ha='center', va='center' if above else 'center',
            fontsize=7, color='#1F3864', wrap=True,
            bbox=dict(boxstyle='round,pad=0.2', facecolor='#EBF5FB', edgecolor='#2E75B6', linewidth=0.5))

ax.set_title('Key Milestones in the History of Artificial Intelligence', 
             fontsize=13, fontweight='bold', color='#1F3864', pad=15)
plt.tight_layout()
plt.savefig('ch01_timeline.png', dpi=150, bbox_inches='tight')
plt.show()
print("Timeline saved to ch01_timeline.png")
```
