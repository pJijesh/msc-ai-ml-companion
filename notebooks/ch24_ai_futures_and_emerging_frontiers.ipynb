## Scaling laws: model performance vs compute


```python
# Scaling laws: model performance vs compute
import numpy as np
import matplotlib.pyplot as plt

# Simplified model of Chinchilla scaling laws
# L(N, D) ≈ E + A/N^alpha + B/D^beta
# N = parameters, D = training tokens, L = loss
# Constants from Hoffmann et al. (2022)

def loss_chinchilla(N, D, E=1.69, A=406.4, B=410.7, alpha=0.34, beta=0.28):
    return E + A / (N**alpha) + B / (D**beta)

# Fix compute C = 6*N*D (approx FLOPs), optimal: N = D (Chinchilla optimal)
C_values = np.logspace(18, 24, 7)   # 10^18 to 10^24 FLOPs
N_range  = np.logspace(7, 12, 200)  # 10M to 1T parameters

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left: loss curves at different compute budgets
colors = plt.cm.viridis(np.linspace(0.1, 0.9, len(C_values)))
for C, col in zip(C_values, colors):
    D_vals = C / (6 * N_range)
    D_vals = np.clip(D_vals, 1e6, None)
    losses = loss_chinchilla(N_range, D_vals)
    axes[0].plot(N_range, losses, color=col, lw=1.5,
                 label=f'C = {C:.0e} FLOPs')

axes[0].set_xscale('log'); axes[0].set_xlabel('Model Parameters (N)')
axes[0].set_ylabel('Training Loss'); axes[0].set_ylim(1.7, 3.5)
axes[0].legend(fontsize=8, title='Compute budget', loc='upper right')
axes[0].set_title('Scaling Laws: Loss vs Parameters
(Chinchilla-style)', fontsize=11)
axes[0].grid(alpha=0.3)

# Right: optimal model size vs compute (Chinchilla frontier)
C_plot = np.logspace(18, 25, 200)
N_opt  = (C_plot / 6) ** 0.5       # Chinchilla: N_opt ≈ sqrt(C/6)
D_opt  = C_plot / (6 * N_opt)

ax2 = axes[1]
ax2.loglog(C_plot, N_opt, color='#2E75B6', lw=2, label='Optimal N (parameters)')
ax2.loglog(C_plot, D_opt, color='#C0392B', lw=2, linestyle='--', label='Optimal D (tokens)')

# Annotate known models (approximate)
models = [
    ("GPT-3
175B", 6*175e9*300e9, 175e9),
    ("Chinchilla
70B", 6*70e9*1.4e12, 70e9),
    ("Llama-2
70B", 6*70e9*2e12, 70e9),
]
for name, C, N in models:
    ax2.scatter(C, N, s=80, zorder=5, color='#1F3864')
    ax2.annotate(name, (C, N), textcoords='offset points', xytext=(10, -5), fontsize=8)

ax2.set_xlabel('Training Compute (FLOPs)'); ax2.set_ylabel('Model / Dataset Size')
ax2.set_title('Chinchilla Scaling: Compute-optimal Frontier
(equal budget → equal params & tokens)', fontsize=11)
ax2.legend(fontsize=9); ax2.grid(alpha=0.3, which='both')

plt.suptitle('Neural Scaling Laws (Hoffmann et al., 2022 — Chinchilla)', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('ch24_scaling_laws.png', dpi=150, bbox_inches='tight')
plt.show()
```
