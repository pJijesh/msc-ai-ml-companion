## McNemar's test: comparing two classifiers on the same test set


```python
# McNemar's test: comparing two classifiers on the same test set
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from scipy.stats import chi2
import numpy as np

X, y = make_classification(n_samples=1000, n_features=20, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

clf_a = LogisticRegression(random_state=42).fit(X_train, y_train)
clf_b = RandomForestClassifier(n_estimators=100, random_state=42).fit(X_train, y_train)

pred_a = clf_a.predict(X_test)
pred_b = clf_b.predict(X_test)

# Contingency table
# n00 = both wrong, n01 = A wrong B right, n10 = A right B wrong, n11 = both right
n00 = np.sum((pred_a != y_test) & (pred_b != y_test))
n01 = np.sum((pred_a != y_test) & (pred_b == y_test))
n10 = np.sum((pred_a == y_test) & (pred_b != y_test))
n11 = np.sum((pred_a == y_test) & (pred_b == y_test))

# McNemar statistic (with continuity correction)
mcnemar_stat = (abs(n01 - n10) - 1)**2 / (n01 + n10)
p_value = 1 - chi2.cdf(mcnemar_stat, df=1)

print(f"Classifier A (Logistic Regression) accuracy: {accuracy_score(y_test, pred_a):.4f}")
print(f"Classifier B (Random Forest) accuracy:       {accuracy_score(y_test, pred_b):.4f}")
print()
print(f"Contingency table: n01={n01}, n10={n10}")
print(f"McNemar statistic: {mcnemar_stat:.4f}")
print(f"p-value: {p_value:.4f}")
print()
if p_value < 0.05:
    print("Difference is statistically significant (p < 0.05)")
else:
    print("No statistically significant difference (p >= 0.05)")
```

## Multiple comparisons: Bonferroni correction


```python
# Multiple comparisons: Bonferroni correction
import numpy as np

# Simulate 20 experiments, each with p-value drawn from uniform if null is true
np.random.seed(42)
n_experiments = 20
alpha = 0.05

# Under the null hypothesis, p-values are uniform [0,1]
p_values = np.random.uniform(0, 1, n_experiments)

# Without correction
false_positives_naive = np.sum(p_values < alpha)
print(f"Without correction: {false_positives_naive}/{n_experiments} experiments significant at alpha={alpha}")

# With Bonferroni correction
alpha_bonferroni = alpha / n_experiments
false_positives_bonf = np.sum(p_values < alpha_bonferroni)
print(f"With Bonferroni:    {false_positives_bonf}/{n_experiments} experiments significant at alpha={alpha_bonferroni:.4f}")

print()
print(f"Expected false positives without correction: {n_experiments * alpha:.1f}")
print(f"P(at least one false positive, no correction): {1 - (1-alpha)**n_experiments:.4f}")
print(f"P(at least one false positive, Bonferroni):    {1 - (1-alpha_bonferroni)**n_experiments:.4f}")
```
