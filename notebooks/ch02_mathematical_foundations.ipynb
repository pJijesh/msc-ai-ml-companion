## NumPy eigenvectors and eigenvalues


```python
# NumPy eigenvectors and eigenvalues
import numpy as np

A = np.array([[3, 1],
              [1, 3]])

eigenvalues, eigenvectors = np.linalg.eig(A)
print("Eigenvalues:", eigenvalues)    # [4. 2.]
print("Eigenvectors:\n", eigenvectors)

# Verify: Av = lambda * v
v0 = eigenvectors[:, 0]
print("Av:", A @ v0)
print("lambda*v:", eigenvalues[0] * v0)
```

## Automatic differentiation with PyTorch


```python
# Automatic differentiation with PyTorch
import torch

x = torch.tensor(3.0, requires_grad=True)
y = x ** 2 + 2 * x + 1  # f(x) = x^2 + 2x + 1

y.backward()             # compute gradients
print(x.grad)            # tensor(8.) = 2*3 + 2
```

## Cross-entropy loss vs MSE: intuition


```python
# Cross-entropy loss vs MSE: intuition
import numpy as np
import matplotlib.pyplot as plt

probs = np.linspace(0.01, 0.99, 200)

# True label y=1
ce_loss  = -np.log(probs)           # cross-entropy: -log(p)
mse_loss = (1 - probs)**2           # MSE: (y - p)^2

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(probs, ce_loss,  color='#2E75B6', lw=2, label='Cross-entropy')
axes[0].plot(probs, mse_loss, color='#C0392B', lw=2, label='MSE', linestyle='--')
axes[0].set_xlabel('Predicted probability for correct class')
axes[0].set_ylabel('Loss')
axes[0].set_title('Loss function comparison (y = 1)')
axes[0].legend()
axes[0].grid(alpha=0.3)

# Gradient comparison
grad_ce  = -1 / probs
grad_mse = -2 * (1 - probs)

axes[1].plot(probs, np.abs(grad_ce),  color='#2E75B6', lw=2, label='|grad CE|')
axes[1].plot(probs, np.abs(grad_mse), color='#C0392B', lw=2, label='|grad MSE|', linestyle='--')
axes[1].set_xlabel('Predicted probability')
axes[1].set_ylabel('|Gradient|')
axes[1].set_title('Gradient magnitude (steepness of learning signal)')
axes[1].set_ylim(0, 10)
axes[1].legend()
axes[1].grid(alpha=0.3)

plt.suptitle('Why cross-entropy is preferred for classification', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('ch02_loss_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
```

## KL divergence demonstration


```python
# KL divergence demonstration
import numpy as np

def kl_divergence(p, q):
    """KL(p || q) — information lost when approximating p with q."""
    p = np.clip(p, 1e-10, 1)
    q = np.clip(q, 1e-10, 1)
    return np.sum(p * np.log(p / q))

# True distribution p and two approximations
p = np.array([0.5, 0.3, 0.15, 0.05])   # true
q_good = np.array([0.45, 0.32, 0.16, 0.07])  # close approximation
q_poor = np.array([0.25, 0.25, 0.25, 0.25])  # uniform (poor approximation)

print(f"KL(p || q_good) = {kl_divergence(p, q_good):.4f}")
print(f"KL(p || q_poor) = {kl_divergence(p, q_poor):.4f}")
print(f"KL(p || p)      = {kl_divergence(p, p):.4f}  (self = 0)")
print()
print("Cross-entropy H(p,q) = H(p) + KL(p||q)")
H_p = -np.sum(p * np.log(p))
print(f"H(p)         = {H_p:.4f}")
print(f"H(p, q_good) = {H_p + kl_divergence(p, q_good):.4f}")
print(f"H(p, q_poor) = {H_p + kl_divergence(p, q_poor):.4f}")
```
