## EU AI Act: risk classification decision tool


```python
# EU AI Act: risk classification decision tool
def classify_ai_system(description: str, outputs_affect_people: bool = True) -> dict:
    """
    Educational decision aid for EU AI Act risk classification.
    NOT legal advice — consult qualified legal counsel for compliance.
    """
    desc_lower = description.lower()
    
    prohibited_keywords = [
        'social scoring', 'mass surveillance', 'subliminal manipulation',
        'exploit vulnerabilities', 'real-time biometric public space',
        'predict political opinions', 'emotion recognition workplace school'
    ]
    
    high_risk_sectors = {
        'biometric': 'Biometric identification/categorisation',
        'critical infrastructure': 'Critical infrastructure management',
        'education': 'Education/vocational training',
        'employment': 'Employment/HR decisions',
        'credit': 'Credit scoring/financial services',
        'insurance': 'Insurance risk assessment',
        'law enforcement': 'Law enforcement',
        'migration': 'Migration/asylum/border control',
        'justice': 'Administration of justice',
        'medical': 'Medical device/healthcare',
        'safety component': 'Safety-critical component',
    }
    
    limited_risk = ['chatbot', 'deepfake', 'emotion recognition consumer',
                    'generates text', 'generates images']
    
    # Check prohibited
    for kw in prohibited_keywords:
        if kw in desc_lower:
            return {
                'tier': 'PROHIBITED',
                'colour': '🔴',
                'reason': f'Matches prohibited practice: "{kw}"',
                'action': 'Cannot be deployed in EU. Prohibited from February 2025.',
                'fines': 'Up to €35M or 7% global annual turnover'
            }
    
    # Check high-risk
    for kw, sector in high_risk_sectors.items():
        if kw in desc_lower and outputs_affect_people:
            return {
                'tier': 'HIGH RISK',
                'colour': '🟠',
                'reason': f'Falls under high-risk sector: {sector}',
                'action': 'Full compliance required by Aug 2026: risk management system, '
                          'data governance, technical documentation, human oversight, '
                          'accuracy requirements, logging, conformity assessment.',
                'fines': 'Up to €15M or 3% global annual turnover for non-compliance'
            }
    
    # Check limited risk
    for kw in limited_risk:
        if kw in desc_lower:
            return {
                'tier': 'LIMITED RISK',
                'colour': '🟡',
                'reason': f'Involves human-AI interaction or synthetic content',
                'action': 'Transparency obligations: disclose AI nature to users, '
                          'label synthetic content.',
                'fines': 'Up to €7.5M or 1.5% for transparency violations'
            }
    
    return {
        'tier': 'MINIMAL RISK',
        'colour': '🟢',
        'reason': 'No specific high-risk characteristics identified',
        'action': 'No mandatory requirements, but voluntary codes of conduct recommended.',
        'fines': 'N/A'
    }

# Test cases
test_cases = [
    ("AI system for employee performance scoring in recruitment", True),
    ("Chatbot for customer service on a retail website", True),
    ("Government system for social credit scoring of citizens", True),
    ("Spam filter for email", False),
    ("Medical diagnostic AI for detecting cancer in X-rays", True),
    ("AI for predicting student exam results for university admissions", True),
    ("AI-generated image creation tool for artists", True),
]

print("EU AI Act Risk Classification Tool")
print("=" * 65)
print("(Educational purposes only — not legal advice)\n")

for description, affects_people in test_cases:
    result = classify_ai_system(description, affects_people)
    print(f"{result['colour']} {result['tier']}")
    print(f"   System: {description[:60]}")
    print(f"   Reason: {result['reason']}")
    print(f"   Action: {result['action'][:80]}...")
    print()
```
