## Dissertation planning: Gantt chart template


```python
# Dissertation planning: Gantt chart template
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

tasks = [
    ("Topic selection & scoping",    1,  2, '#2E75B6'),
    ("Systematic literature review", 2,  6, '#1F7D6E'),
    ("Research proposal",            4,  5, '#5B2C9E'),
    ("Ethics approval (if required)",4,  5, '#C7770A'),
    ("Data collection / experiments",5, 10, '#C0392B'),
    ("Model development",            6, 11, '#C0392B'),
    ("Results analysis",             10,13, '#2E75B6'),
    ("Writing: Methods chapter",     7, 11, '#1F3864'),
    ("Writing: Introduction",        8, 12, '#1F3864'),
    ("Writing: Results & Discussion",11,14, '#1F3864'),
    ("First draft submission",       14,14, '#888888'),
    ("Supervisor feedback & revision",14,16,'#5B2C9E'),
    ("Final submission",             16,17, '#1F7D6E'),
    ("Viva preparation",             16,18, '#C7770A'),
]

fig, ax = plt.subplots(figsize=(14, 8))
months = range(1, 19)
ax.set_xlim(0, 19)
ax.set_ylim(-1, len(tasks))
ax.set_xlabel("Month", fontsize=11)
ax.set_title("MSc Dissertation: 18-Month Project Gantt Chart Template", fontsize=13, fontweight='bold')
ax.set_xticks(months)
ax.set_xticklabels([f"M{m}" for m in months], fontsize=9)
ax.set_yticks(range(len(tasks)))
ax.set_yticklabels([t[0] for t in tasks], fontsize=9)
ax.grid(axis='x', alpha=0.3)

for i, (name, start, end, color) in enumerate(tasks):
    ax.barh(i, end - start + 0.8, left=start - 0.4, height=0.6,
            color=color, alpha=0.8, edgecolor='white', linewidth=0.5)
    if end > start:
        ax.text((start + end) / 2, i, f'M{start}–M{end}',
                ha='center', va='center', fontsize=7, color='white', fontweight='bold')

# Key milestones
milestones = [(5,"Proposal
approved"), (14,"First
draft"), (16,"Final
submission")]
for m, label in milestones:
    ax.axvline(x=m, color='#C7770A', lw=1.5, linestyle='--', alpha=0.7)
    ax.text(m, len(tasks)-0.3, label, ha='center', fontsize=7.5, color='#C7770A', fontweight='bold')

ax.invert_yaxis()
plt.tight_layout()
plt.savefig('ch23_gantt.png', dpi=150, bbox_inches='tight')
plt.show()
print("Tip: Adjust this template to your programme's specific timeline and deadlines.")
```
