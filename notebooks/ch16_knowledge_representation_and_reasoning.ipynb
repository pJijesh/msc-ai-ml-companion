## Building and querying a simple knowledge graph with NetworkX + RDFLib


```python
# Building and querying a simple knowledge graph with NetworkX + RDFLib
# Install: pip install networkx rdflib matplotlib

import networkx as nx
import matplotlib.pyplot as plt

# Build a simple RDF-style knowledge graph as a directed graph
G = nx.DiGraph()

triples = [
    ("Turing",       "nationality",    "British"),
    ("Turing",       "invented",       "Turing Test"),
    ("Turing",       "worked_at",      "Bletchley Park"),
    ("Hinton",       "nationality",    "British"),
    ("Hinton",       "invented",       "Backpropagation"),
    ("Hinton",       "won",            "Turing Award"),
    ("LeCun",        "nationality",    "French"),
    ("LeCun",        "invented",       "ConvNet"),
    ("LeCun",        "won",            "Turing Award"),
    ("Bengio",       "nationality",    "Canadian"),
    ("Bengio",       "invented",       "Attention Mechanism"),
    ("Bengio",       "won",            "Turing Award"),
    ("Turing Award", "awarded_by",     "ACM"),
    ("Backpropagation", "enables",     "Deep Learning"),
    ("ConvNet",      "enables",        "Computer Vision"),
    ("Attention Mechanism", "enables", "Transformers"),
    ("Transformers", "enables",        "LLMs"),
]

for s, p, o in triples:
    G.add_edge(s, o, predicate=p)

# Simple query: "Who won the Turing Award?"
turing_award_winners = [s for s, o, d in G.edges(data=True) if o == "Turing Award"]
print("Who won the Turing Award?")
for w in turing_award_winners:
    print(f"  → {w}")

# Path query: how does Hinton connect to LLMs?
try:
    path = nx.shortest_path(G, "Hinton", "LLMs")
    print("\nPath from Hinton to LLMs:")
    for i in range(len(path)-1):
        pred = G[path[i]][path[i+1]]['predicate']
        print(f"  {path[i]} --[{pred}]--> {path[i+1]}")
except nx.NetworkXNoPath:
    print("No path found")

# Visualise
plt.figure(figsize=(14, 9))
pos = nx.spring_layout(G, seed=42, k=2.5)
node_colors = []
for n in G.nodes():
    if n in ["Turing","Hinton","LeCun","Bengio"]: node_colors.append('#2E75B6')
    elif n in ["Turing Award","Turing Test"]:       node_colors.append('#C7770A')
    elif n in ["Deep Learning","LLMs","Transformers","Computer Vision"]: node_colors.append('#1F7D6E')
    else: node_colors.append('#D9DCE0')

nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=1000, alpha=0.9)
nx.draw_networkx_labels(G, pos, font_size=8, font_weight='bold')
edge_labels = nx.get_edge_attributes(G, 'predicate')
nx.draw_networkx_edges(G, pos, arrows=True, arrowsize=20, alpha=0.6)
nx.draw_networkx_edge_labels(G, pos, edge_labels, font_size=7, alpha=0.8)
plt.title('AI Knowledge Graph: entities and relationships', fontsize=13, fontweight='bold')
plt.axis('off'); plt.tight_layout()
plt.savefig('ch16_knowledge_graph.png', dpi=150, bbox_inches='tight')
plt.show()
```
